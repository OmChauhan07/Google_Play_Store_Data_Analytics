## Importing Libraries

```python
import numpy as np
import pandas as pd
```

## Importing PlayStore App Data

```python
app_data = pd.read_csv('..\Data\Raw_Data\googleplaystore.csv')
app_data.head()
```

## Creating Copy of the Raw Dataset

```python
app_df = app_data.copy()
app_df.info()
```

```python
# Size of the dataset
print("Rows and Columns:", app_df.shape)
```

```python
app_df.describe()
```

## Fixing Null Values

```python
# Checking for missing values
app_df.isnull().sum()
```

```python
# Replacing Rating Null values with Median
app_df['Rating'].fillna(app_df['Rating'].median(), inplace=True)

# Dropping rows with null values in Type and Content Rating
app_df = app_df.dropna(subset=["Type", "Content Rating"])

# Replacing null values in Current Ver and Android Ver with Unknown
app_df['Current Ver'].fillna('Unknown', inplace=True)
app_df['Android Ver'].fillna('Unknown', inplace=True)
```

```python
# Final check for missing values
app_df.isna().sum()
```

```python
# Validating that there are no more missing values
app_df[app_df["Rating"].isna()]
```

```python
# Dropping any remaining rows with null values in Rating, Current Ver and Android Ver
app_df = app_df.dropna(subset=["Rating", "Current Ver", "Android Ver"])
```

```python
# Validating that there are no more missing values
app_df.isna().sum()
```

```python
# checking New size of the dataset
app_df.shape
```

## Fixing Duplicate Values

```python
# Checking for duplicates
app_df.duplicated().sum()
```

```python
# Identifying duplicate entries based on the App column
app_df[app_df["App"].duplicated(keep=False)].sort_values(by="App")
```

```python
# Sorting by number of reviews and removing duplicates
app_df = app_df.sort_values(by="Reviews", ascending=False)

# Dropping duplicates based on the App column, keeping the first occurrence
app_df = app_df.drop_duplicates(subset="App", keep="first")

# Final check for duplicates based on the App column
print(app_df["App"].duplicated().sum())
```

```python
app_df.shape
```

```python
app_df.info()
```

## Cleaning Size Column

```python
# Checking unique values in the Size column
app_df["Size"].unique()[:20]
```

```python
# Create a function to clean the Size column
def clean_size(size):
    if size == "Varies with device":
        return np.nan
    
    size = str(size)

    if "M" in size:
        return float(size.replace("M", ""))

    elif "k" in size:
        return float(size.replace("k", "")) / 1024

    else:
        return np.nan
```

```python
# Applying the clean_size function to the Size column
app_df["Size"] = app_df["Size"].apply(clean_size)
```

```python
# Checking the cleaned Size column
app_df["Size"] = app_df["Size"].fillna(app_df["Size"].median())
```

```python
# Final check of the Size column
app_df["Size"].head()
```

## Clean Install Column

```python
# Checking unique values in the Installs column
app_df["Installs"].unique()[:10]
```

```python
# Cleaning the Installs column by removing + and , elements
app_df["Installs"] = (
    app_df["Installs"]
    .str.replace("+", "", regex=False)
    .str.replace(",", "", regex=False)
)
```

```python
# Converting the Installs column to numeric, coercing errors to NaN
app_df["Installs"] = pd.to_numeric(
    app_df["Installs"],
    errors="coerce"
)
```

```python
# Final check of the Installs column
app_df["Installs"].head()
```

## Cleaning Price Column

```python
# Checking unique values in the Price column
app_df["Price"] = (
    app_df["Price"]
    .str.replace("$", "", regex=False)
)
```

```python
# Converting the Price column to numeric, coercing errors to NaN
app_df["Price"] = pd.to_numeric(
    app_df["Price"],
    errors="coerce"
)
```

```python
# Final check of the Price column
app_df["Price"].head()
```

## Cleaning Review Column

```python
# Checking unique values in the Reviews column
app_df["Reviews"] = pd.to_numeric(
    app_df["Reviews"],
    errors="coerce"
)
```

```python
# Final check of the Reviews column
app_df["Reviews"].head()
```

## Cleaning Lastly updated Column

```python
# Checking unique values in the Last Updated column
app_df["Last Updated"] = pd.to_datetime(
    app_df["Last Updated"],
    errors="coerce"
)
```

```python
# Final check of the Last Updated column
app_df["Last Updated"].head()
```

## Feature Engineering

```python
# Extracting month and year from the Last Updated column
app_df["Updated_Month"] = app_df["Last Updated"].dt.month

# Extracting year from the Last Updated column
app_df["Updated_Year"] = app_df["Last Updated"].dt.year

# Final check of the Last Updated, Updated_Month and Updated_Year columns
app_df[["Last Updated", "Updated_Month", "Updated_Year"]].head()
```

## Final Check over the Dataset
