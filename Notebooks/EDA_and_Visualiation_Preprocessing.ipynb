## Category Analysis

```python
# Top 10 categories by count
df["Category"].value_counts().head(10)
```

```python
# to visualize the top 10 categories by count
top_categories = df["Category"].value_counts().head(10)

# Size of the plot
plt.figure(figsize=(12, 6))

# Bar plot using seaborn
sns.barplot(
    x=top_categories.index,
    y=top_categories.values
)

# format the labels for better readability
plt.xticks(rotation=45)
plt.title("Top 10 App Categories by Count")
plt.xlabel("Category")
plt.ylabel("Number of Apps")  

plt.show()  
```

```python
# Top 10 categories by total installs
category_installs = (
    df.groupby("Category")["Installs"].sum().sort_values(ascending=False).head(10)
)

category_installs
```

```python
plt.figure(figsize=(12, 6))

ax = sns.barplot(
    x=category_installs.index,
    y=category_installs.values
)

plt.xticks(rotation=45)

plt.title("Top Categories by Installs")
plt.xlabel("Category")
plt.ylabel("Total Installs")

# Y-axis formatting
ax.set_yticklabels([
    f'{x/1e9:.1f}B' if x >= 1e9
    else f'{x/1e6:.1f}M'
    for x in ax.get_yticks()
])

# Value labels
for i, value in enumerate(category_installs.values):

    label = (
        f'{value/1e9:.1f}B'
        if value >= 1e9
        else f'{value/1e6:.1f}M'
    )

    ax.text(
        i,
        value,
        label,
        ha='center',
        va='bottom'
    )

plt.show()
```

```python
# average rating by category
avg_rating = (
    df.groupby("Category")["Rating"].mean().sort_values(ascending=False).head(10)
)

avg_rating
```

```python
plt.figure(figsize=(12, 6))

sns.barplot(
    x=avg_rating.index,
    y=avg_rating.values
)

plt.xticks(rotation=45)
plt.title("Top 10 App Categories by Average Rating")
plt.xlabel("Category")
plt.ylabel("Average Rating")
plt.show()
```
