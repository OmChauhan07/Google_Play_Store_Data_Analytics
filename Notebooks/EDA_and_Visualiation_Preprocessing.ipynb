## Outlier Detection

```python
plt.figure(figsize=(12, 6))

sns.boxplot(
    x = df["Rating"]
)

plt.title("Box Plot of App Ratings")
plt.show()
```

```python
plt.figure(figsize=(10, 5))

ax = sns.boxplot(
    x=df["Installs"]
)

# Formatter function
def human_format(x, pos):
    if x >= 1e12:
        return f'{x/1e12:.1f}T'
    elif x >= 1e9:
        return f'{x/1e9:.1f}B'
    elif x >= 1e6:
        return f'{x/1e6:.1f}M'
    elif x >= 1e3:
        return f'{x/1e3:.1f}K'
    return str(x)

# Apply formatter
ax.xaxis.set_major_formatter(
    FuncFormatter(human_format)
)

plt.title("Installs Outliers")
plt.show()
```

```python
plt.figure(figsize=(10, 5))

ax = sns.boxplot(
    x=df["Revenue"]
)

# Formatter function
def human_format(x, pos):
    if x >= 1e12:
        return f'${x/1e12:.1f}T'
    elif x >= 1e9:
        return f'${x/1e9:.1f}B'
    elif x >= 1e6:
        return f'${x/1e6:.1f}M'
    elif x >= 1e3:
        return f'${x/1e3:.1f}K'
    return f'${x:.0f}'

# Apply formatter
ax.xaxis.set_major_formatter(
    FuncFormatter(human_format)
)

plt.title("Revenue Outliers")
plt.show()
```

## Category-wise Time Trend Analysis

```python
# Year Month Columns
df["Year_Month"] = (
    df["Last Updated"]
    .dt.to_period("M")
    .astype(str)
)
```

```python
# Monthly Installs by Category
monthly_category_installs = (
    df.groupby(
        ["Year_Month", "Category"]
    )["Installs"]
    .sum()
    .reset_index()
)

monthly_category_installs.head()
```

```python
# Top Categories for Trend Plot
top_categories = (
    df.groupby("Category")["Installs"]
    .sum()
    .sort_values(ascending=False)
    .head(5)
    .index
)

top_categories
```

```python
trend_df = monthly_category_installs[
    monthly_category_installs["Category"]
    .isin(top_categories)
]
```

```python
# Time Series Plot
plt.figure(figsize=(14, 6))

sns.lineplot(
    data=trend_df,
    x="Year_Month",
    y="Installs",
    hue="Category",
    marker="o"
)

plt.title("Monthly Installs Trend by Category")

plt.xlabel("Year-Month")
plt.ylabel("Installs")

plt.xticks(rotation=45)

plt.show()
```

## Month-over-Month Growth Preparation

```python
# Sort Data Properly
trend_df = trend_df.sort_values(
    by=["Category", "Year_Month"]
)
```

```python
# Category Growth Percentage
trend_df["MoM_Growth"] = (
    trend_df.groupby("Category")["Installs"]
    .pct_change() * 100
)
```

```python
trend_df.head(15)
```

## Growth Visualization

```python
plt.figure(figsize=(14, 6))

sns.lineplot(
    data=trend_df,
    x="Year_Month",
    y="MoM_Growth",
    hue="Category",
    marker="o"
)

plt.title("Month-over-Month Growth by Category")

plt.xlabel("Year-Month")
plt.ylabel("Growth %")

plt.xticks(rotation=45)

plt.axhline(
    y=20,
    color='red',
    linestyle='--',
    label='20% Growth'
)

plt.legend()

plt.show()
```
