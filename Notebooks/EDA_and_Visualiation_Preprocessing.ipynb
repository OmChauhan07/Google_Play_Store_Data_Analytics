## Category Analysis

```python
# Top 10 categories by count
df["Category"].value_counts().head(10)
```

```python
# to visualize the top 10 categories by count
top_categories = df["Category"].value_counts().head(10)

# Size of the plot
plt.figure(figsize=(12, 6))

# Bar plot using seaborn
sns.barplot(
    x=top_categories.index,
    y=top_categories.values
)

# format the labels for better readability
plt.xticks(rotation=45)
plt.title("Top 10 App Categories by Count")
plt.xlabel("Category")
plt.ylabel("Number of Apps")  

plt.show()  
```

```python
# Top 10 categories by total installs
category_installs = (
    df.groupby("Category")["Installs"].sum().sort_values(ascending=False).head(10)
)

category_installs
```

```python
plt.figure(figsize=(12, 6))

ax = sns.barplot(
    x=category_installs.index,
    y=category_installs.values
)

plt.xticks(rotation=45)

plt.title("Top Categories by Installs")
plt.xlabel("Category")
plt.ylabel("Total Installs")

# Y-axis formatting
ax.set_yticklabels([
    f'{x/1e9:.1f}B' if x >= 1e9
    else f'{x/1e6:.1f}M'
    for x in ax.get_yticks()
])

# Value labels
for i, value in enumerate(category_installs.values):

    label = (
        f'{value/1e9:.1f}B'
        if value >= 1e9
        else f'{value/1e6:.1f}M'
    )

    ax.text(
        i,
        value,
        label,
        ha='center',
        va='bottom'
    )

plt.show()
```

```python
# average rating by category
avg_rating = (
    df.groupby("Category")["Rating"].mean().sort_values(ascending=False).head(10)
)

avg_rating
```

```python
plt.figure(figsize=(12, 6))

sns.barplot(
    x=avg_rating.index,
    y=avg_rating.values
)

plt.xticks(rotation=45)
plt.title("Top 10 App Categories by Average Rating")
plt.xlabel("Category")
plt.ylabel("Average Rating")
plt.show()
```

```python
category_revenue = (
    df.groupby("Category")["Revenue"].sum().sort_values(ascending=False).head(10)
)

category_revenue
```

```python
plt.figure(figsize=(12, 6))

ax = sns.barplot(
    x=category_revenue.index,
    y=category_revenue.values
)

plt.xticks(rotation=45)

plt.title("Top Categories by Revenue")
plt.xlabel("Category")
plt.ylabel("Total Revenue")

# Format Y-axis labels
ax.set_yticklabels([
    f'${x/1e12:.1f}T'
    if x >= 1e12
    else f'${x/1e9:.1f}B'
    if x >= 1e9
    else f'${x/1e6:.1f}M'
    if x >= 1e6
    else f'${x/1e3:.1f}K'
    for x in ax.get_yticks()
])

# Add labels above bars
for i, value in enumerate(category_revenue.values):

    label = (
        f'${value/1e12:.1f}T'
        if value >= 1e12
        else f'${value/1e9:.1f}B'
        if value >= 1e9
        else f'${value/1e6:.1f}M'
        if value >= 1e6
        else f'${value/1e3:.1f}K'
    )

    ax.text(
        i,
        value,
        label,
        ha='center',
        va='bottom'
    )

plt.show()
```

## Correlation Analysis

```python
correlation_cols = [
    "Rating",
    "Reviews",
    "Size",
    "Installs",
    "Price",
    "Revenue",
    "Avg_Sentiment_Polarity",
    "Avg_Sentiment_Subjectivity"
]

corr_matrix = df[correlation_cols].corr()

corr_matrix
```

```python
plt.figure(figsize=(10, 7))

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")

plt.show()
```

## Outlier Detection

```python
plt.figure(figsize=(12, 6))

sns.boxplot(
    x = df["Rating"]
)

plt.title("Box Plot of App Ratings")
plt.show()
```

```python
plt.figure(figsize=(10, 5))

ax = sns.boxplot(
    x=df["Installs"]
)

# Formatter function
def human_format(x, pos):
    if x >= 1e12:
        return f'{x/1e12:.1f}T'
    elif x >= 1e9:
        return f'{x/1e9:.1f}B'
    elif x >= 1e6:
        return f'{x/1e6:.1f}M'
    elif x >= 1e3:
        return f'{x/1e3:.1f}K'
    return str(x)

# Apply formatter
ax.xaxis.set_major_formatter(
    FuncFormatter(human_format)
)

plt.title("Installs Outliers")
plt.show()
```

```python
plt.figure(figsize=(10, 5))

ax = sns.boxplot(
    x=df["Revenue"]
)

# Formatter function
def human_format(x, pos):
    if x >= 1e12:
        return f'${x/1e12:.1f}T'
    elif x >= 1e9:
        return f'${x/1e9:.1f}B'
    elif x >= 1e6:
        return f'${x/1e6:.1f}M'
    elif x >= 1e3:
        return f'${x/1e3:.1f}K'
    return f'${x:.0f}'

# Apply formatter
ax.xaxis.set_major_formatter(
    FuncFormatter(human_format)
)

plt.title("Revenue Outliers")
plt.show()
```

## Time Based Feature Extraction

```python
df["Last Updated"] = pd.to_datetime(
    df["Last Updated"],
    errors="coerce"
)
```

```python
df["Update_Month_Name"] = (
    df["Last Updated"]
    .dt.month_name()
)
```

```python
df["Update_Year"] = (
    df["Last Updated"]
    .dt.year
)
```

```python
df[
    [
        "Last Updated",
        "Update_Month_Name",
        "Update_Year"
    ]
].head()
```

```python
monthly_installs = (
    df.groupby("Update_Month_Name")["Installs"]
    .sum()
)

monthly_installs
```

```python
plt.figure(figsize=(12, 6))

sns.lineplot(
    x=monthly_installs.index,
    y=monthly_installs.values,
    marker="o"
)

plt.title("Monthly Installs Trend")
plt.xlabel("Month")
plt.ylabel("Total Installs")

plt.xticks(rotation=45)

plt.show()
```
