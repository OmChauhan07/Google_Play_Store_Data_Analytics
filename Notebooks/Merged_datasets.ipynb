## Importing Libraries

```python
import numpy as np
import pandas as pd
```

## Importing Datasets

```python
# Merging the two datasets
app_df = pd.read_csv('..//Data//Cleaned_Data//cleaned_app_data.csv')
review_df = pd.read_csv('..//Data//Cleaned_Data//cleaned_review_data.csv')
```

```python
print(app_df.head())
print(review_df.head())
```

## Merging Full Dataset (Contains Multiple row of Same App)

```python
# performing a left join to merge the datasets on the App column
merged_df = pd.merge(app_df, review_df, on='App', how='left')
```

```python
# validating the merged dataset
print(merged_df.shape)
merged_df.head()
```

```python
merged_df.to_csv('..\\Data\\Cleaned_Data\\merged_dataset.csv', index=False)
```

## Aggregated Dataset (Contains one row per each app)

```python
review_df.info()
```

```python
# Group Review Dataset
review_agg = review_df.groupby("App").agg({
    "Sentiment_Polarity": "mean",
    "Sentiment_Subjectivity": "mean",
    "Translated_Review": "count",
    "Review_length": "mean",
    "Word_count": "mean"
}).reset_index()
```

```python
# renaming columns for clarity
review_agg = review_agg.rename(columns={
    "Translated_Review": "Total_Reviews",
    "Sentiment_Polarity": "Avg_Sentiment_Polarity",
    "Sentiment_Subjectivity": "Avg_Sentiment_Subjectivity",
    "Review_length": "Avg_Review_Length",
    "Word_count": "Avg_Word_Count"
})
```

```python
review_agg.head()
```

```python
# Merging Aggregated Review Data with App Data
aggre_df = pd.merge(app_df, review_agg, on='App', how='left')
```

```python
print(aggre_df.shape)
aggre_df.head()
```

```python
# Validating the changes made to the dataset
print(aggre_df.info())
```

```python
aggre_df.isna().sum()
```

```python
review_cols = [
    "Avg_Sentiment_Polarity",
    "Avg_Sentiment_Subjectivity",
    "Total_Reviews",
    "Avg_Review_Length",
    "Avg_Word_Count"
]

aggre_df[review_cols] = aggre_df[review_cols].fillna(0)
```

```python
aggre_df["Revenue"] = (aggre_df["Installs"] * aggre_df["Price"])
```
