## Importing Libraries

```python
import pandas as pd
import numpy as np

import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

import pytz
from datetime import datetime
```

```python
# IST Time Function
def get_ist_hour():
    india_tz = pytz.timezone("Asia/Kolkata")
    current_time = datetime.now(india_tz)
    return current_time.hour
```

```python
# Time Visibility Function
def is_time_between(start_hour, end_hour):
    current_hour = get_ist_hour()
    return start_hour <= current_hour < end_hour
```

```python
# Google's Color Palette
GOOGLE_COLORS = {
    "blue": "#4285F4",
    "green": "#34A853",
    "yellow": "#FBBC05",
    "red": "#EA4335"
}
```

### Task-1: Grouped Bar Chart

```python
t1_df = pd.read_csv("../Data/Task_wise_dataset/task1_grouped.csv")

t1_df
```

```python
t1_df.shape
```

```python
# Group Bar Chart
fig = make_subplots(
    specs=[[{"secondary_y": True}]]
)

# Avg Rating Bar
fig.add_trace(
    go.Bar(
        x=t1_df["Category"],
        y=t1_df["Rating"],
        name="Average Rating",
        marker_color=GOOGLE_COLORS["yellow"],
        hovertemplate="<b>Category:</b> %{x}<br>" + "<b>Average Rating:</b> %{y:.2f}<extra></extra>",
        offsetgroup=1,
        width=0.35
    ),
    secondary_y=False
)

# Total Review Bar
fig.add_trace(
    go.Bar(
        x=t1_df["Category"],
        y=t1_df["Reviews"],
        name="Total Reviews",
        marker_color=GOOGLE_COLORS["blue"],
        hovertemplate="<b>Category:</b> %{x}<br>" + "<b>Total Reviews:</b> %{y:,}<extra></extra>",
        offsetgroup=2,
        width=0.35
    ),
    secondary_y=True
)
```

```python
# Update Layout
fig.update_layout(

    title={
        "text":"Top 10 App Categories: Average Rating vs Total Reviews",
        "x": 0.5
    },

    template="plotly_white",
    barmode="group",
    height=650,
    width=1200,
    legend_title="Metrics",
)
```

```python
# Improving Axis & Hover Style

# X-axis
fig.update_xaxes(
    title_text="Category",
    tickangle=-30
)

# Left Y-axis
fig.update_yaxes(
    title_text="Average Rating",
    secondary_y=False
)

# Right Y-axis
fig.update_yaxes(
    title_text="Total Reviews",
    separatethousands=True,
    secondary_y=True
)
```

```python
fig.show(renderer="browser")
```

### Task-2: Choropleth Map

```python
t2_df = pd.read_csv("../Data/Task_wise_dataset/task2_grouped.csv")
t2_df
```

```python
# Dummy Countries
t2_df["Country"] = [
    "United States",
    "Russia",          
    "Brazil",  
    "India",          
    "Canada"  
]
```

```python
# Layout Styling
fig.update_layout(
    template="plotly_white",
    height=650,
    width=1200,
    title={
        "x": 0.5
    },
    geo=dict(
        showframe=False,
        showcoastlines=True,
        projection_type="equirectangular"
    ),
    coloraxis_colorbar=dict(
        title="Installs"
    )
)
```

```python
fig.show(renderer="browser")
```

### Task-3: Dual-axis Chart

```python
t3_df = pd.read_csv("../Data/Task_wise_dataset/task3_grouped.csv")
t3_df
```

```python
# Create Subplots
fig = make_subplots(
    specs=[[{"secondary_y": True}]]
)
```

```python
# add avg installs line
for app_type, color in zip(
    ["Free", "Paid"],
    [
        GOOGLE_COLORS["green"],
        GOOGLE_COLORS["yellow"]
    ]
):
    filtered_df = t3_df[
        t3_df["Type"] == app_type
    ]
    fig.add_trace(
        go.Bar(
            x=filtered_df["Category"],
            y=filtered_df["Installs"],
            name=f"{app_type} Avg Installs",
            marker_color=color,
            offsetgroup=app_type,
            hovertemplate=
            "<b>Category:</b> %{x}<br>" +
            "<b>Avg Installs:</b> %{y:,}<extra></extra>"
        ),
        secondary_y=False
    )
```

```python
# Added Revenue line
for app_type, color in zip(
    ["Free", "Paid"],
    [
        GOOGLE_COLORS["red"],
        GOOGLE_COLORS["blue"]
    ]
):
    filtered_df = t3_df[
        t3_df["Type"] == app_type
    ]
    fig.add_trace(
        go.Scatter(
            x=filtered_df["Category"],
            y=filtered_df["Revenue"],
            name=f"{app_type} Avg Revenue",
            mode="lines+markers",
            line=dict(
                color=color,
                width=4
            ),
            marker=dict(
                size=10
            ),
            hovertemplate=
            "<b>Category:</b> %{x}<br>" +
            "<b>Avg Revenue:</b> $%{y:,.2f}<extra></extra>"
        ),
        secondary_y=True
    )
```

```python
# Layout Styling
fig.update_layout(
    title={
        "text":
        "Free vs Paid Apps: Average Installs and Revenue",
        "x": 0.5
    },
    template="plotly_white",
    height=650,
    width=1200,
    barmode="group",
    legend_title="Metrics",
    hovermode="x unified"
)
```

```python
# Axis Styling
fig.update_xaxes(
    title_text="Category",
    tickangle=-20
)
# Left Axis
fig.update_yaxes(
    title_text="Average Installs",
    separatethousands=True,
    secondary_y=False
)
# Right Axis
fig.update_yaxes(
    title_text="Average Revenue ($)",
    separatethousands=True,
    secondary_y=True
)
```

```python
fig.show(renderer="browser")
```

### Task-4: Time Series Chart

```python
t4_df = pd.read_csv("../Data/Task_wise_dataset/task4_grouped.csv")
t4_df.head()
```

```python
# Translate Categories
t4_df["Category"] = (
    t4_df["Category"]
    .replace({

        "Beauty": "सौंदर्य",

        "Business": "வணிகம்",

        "Dating": "Partnersuche"
    })
)
```

```python
top_categories = (
    t4_df.groupby("Category")["Installs"]
    .sum()
    .sort_values(ascending=False)
    .head(4)
    .index
)

t4_df = t4_df[
    t4_df["Category"].isin(top_categories)
]
```

```python
# Create Subplots
fig = go.Figure()
```

```python
# Google Colors
colors = [

    GOOGLE_COLORS["blue"],
    GOOGLE_COLORS["green"],
    GOOGLE_COLORS["yellow"],
    GOOGLE_COLORS["red"]
]
```

```python
# add time series lines
for category, color in zip(
    t4_df["Category"].unique(),
    colors * 5
):
    category_df = t4_df[
        t4_df["Category"] == category
    ]

    # Main Trend Line
    fig.add_trace(
        go.Scatter(
            x=category_df["Year_Month"],
            y=category_df["Installs"],
            mode="lines",
            name=category,
            legendgroup=category,
            line=dict(
                color=color,
                width=3
            ),
            hovertemplate=
            "<b>Category:</b> %{fullData.name}<br>" +
            "<b>Month:</b> %{x}<br>" +
            "<b>Total Installs:</b> %{y:,}<extra></extra>"
        )
    )
```

```python
# Higher Growth area
growth_df = category_df[
        category_df["MoM_Growth"] > 20
    ]
```

```python
# add markers
fig.add_trace(
        go.Scatter(
            x=growth_df["Year_Month"],
            y=growth_df["Installs"],
            mode="markers",
            legendgroup=category,
            showlegend=False,
            marker=dict(
                size=14,
                color="white",
                line=dict(
                    color=color,
                    width=4
                ),
                symbol="circle"
            ),
            hovertemplate=
            "<b>High Growth Period</b><br>" +
            "<b>Category:</b> " + category + "<br>" +
            "<b>Month:</b> %{x}<br>" +
            "<b>Total Installs:</b> %{y:,}<extra></extra>"
        )
    )
```

```python
# layout Styling
fig.update_layout(
    title={
        "text":
        "Top 4 App Categories: Total Installs Trend Over Time",
        "x": 0.5
    },
    template="plotly_white",
    height=700,
    width=1250,
    hovermode="x unified",
    legend_title="Categories"
)
```

```python
# X-axis Styling
fig.update_xaxes(
    title_text="Year-Month",
    tickangle=-45
)

# Y-axis Styling
fig.update_yaxes(
    title_text="Total Installs (Log Scale)",
    separatethousands=True,
    type="log"
)
```

```python
fig.show(renderer="browser")
```

### Task-6: Stacked Area Chart

```python
t6_df = pd.read_csv("../Data/Task_wise_dataset/task6_grouped.csv")
t6_df.head()
```

```python
# Traslate Categories
t6_df["Category"] = (
    t6_df["Category"]
    .replace({
        "TRAVEL_AND_LOCAL": "Voyage_Local",
        "PRODUCTIVITY": "Productividad",
        "PHOTOGRAPHY": "写真"
    })
)
```

```python
# Create Figure
fig = go.Figure()
```

```python
# Color Palette
colors = [
    GOOGLE_COLORS["blue"],
    GOOGLE_COLORS["green"],
    GOOGLE_COLORS["yellow"],
    GOOGLE_COLORS["red"],
    "#9C27B0",
    "#00ACC1",
    "#FF7043"
]
```

```python
# Stacked Area Traces
for category, color in zip(
    t6_df["Category"].unique(),
    colors * 5
):
    category_df = t6_df[
        t6_df["Category"] == category
    ]
    fig.add_trace(
        go.Scatter(
            x=category_df["Year_Month"],
            y=category_df["Cumulative_Installs"],
            mode="lines",
            name=category,
            stackgroup="one",
            line=dict(
                width=2,
                color=color
            ),
            hovertemplate=
            "<b>Category:</b> %{fullData.name}<br>" +
            "<b>Month:</b> %{x}<br>" +
            "<b>Cumulative Installs:</b> %{y:,}<extra></extra>"
        )
    )
```

```python
# High Growth Periods
growth_df = t6_df[
    t6_df["MoM_Growth"] > 25
]
```

```python
# Markers
fig.add_trace(
    go.Scatter(
        x=growth_df["Year_Month"],
        y=growth_df["Cumulative_Installs"],
        mode="markers",
        marker=dict(
            size=14,
            color="white",
            line=dict(
                color="black",
                width=3
            ),
            symbol="circle"
        ),
        name="Growth > 25%",
        hovertemplate=
        "<b>High Growth Period</b><br>" +
        "<b>Month:</b> %{x}<br>" +
        "<b>Installs:</b> %{y:,}<extra></extra>"
    )
)
```

```python
# Layout Styling
fig.update_layout(
    title={
        "text":
        "Cumulative Installs by App Category Over Time",
        "x": 0.5
    },
    template="plotly_white",
    height=700,
    width=1250,
    hovermode="x unified",
    legend_title="Categories"
)
```

```python
# Axis Styling
fig.update_xaxes(
    title_text="Year-Month",
    tickangle=-45
)
fig.update_yaxes(
    title_text="Cumulative Installs",
    separatethousands=True
)
```

```python
fig.show(renderer="browser")
```
