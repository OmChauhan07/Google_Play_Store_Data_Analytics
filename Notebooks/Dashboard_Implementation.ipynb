## Importing Libraries

```python
import pandas as pd
import numpy as np

import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

import pytz
from datetime import datetime
```

```python
# IST Time Function
def get_ist_hour():
    india_tz = pytz.timezone("Asia/Kolkata")
    current_time = datetime.now(india_tz)
    return current_time.hour
```

```python
# Time Visibility Function
def is_time_between(start_hour, end_hour):
    current_hour = get_ist_hour()
    return start_hour <= current_hour < end_hour
```

```python
# Google's Color Palette
GOOGLE_COLORS = {
    "blue": "#4285F4",
    "green": "#34A853",
    "yellow": "#FBBC05",
    "red": "#EA4335"
}
```

### Task-1: Grouped Bar Chart

```python
t1_df = pd.read_csv("../Data/Task_wise_dataset/task1_grouped.csv")

t1_df
```

```python
t1_df.shape
```

```python
# Group Bar Chart
fig = make_subplots(
    specs=[[{"secondary_y": True}]]
)

# Avg Rating Bar
fig.add_trace(
    go.Bar(
        x=t1_df["Category"],
        y=t1_df["Rating"],
        name="Average Rating",
        marker_color=GOOGLE_COLORS["yellow"],
        hovertemplate="<b>Category:</b> %{x}<br>" + "<b>Average Rating:</b> %{y:.2f}<extra></extra>",
        offsetgroup=1,
        width=0.35
    ),
    secondary_y=False
)

# Total Review Bar
fig.add_trace(
    go.Bar(
        x=t1_df["Category"],
        y=t1_df["Reviews"],
        name="Total Reviews",
        marker_color=GOOGLE_COLORS["blue"],
        hovertemplate="<b>Category:</b> %{x}<br>" + "<b>Total Reviews:</b> %{y:,}<extra></extra>",
        offsetgroup=2,
        width=0.35
    ),
    secondary_y=True
)
```

```python
# Update Layout
fig.update_layout(

    title={
        "text":"Top 10 App Categories: Average Rating vs Total Reviews",
        "x": 0.5
    },

    template="plotly_white",
    barmode="group",
    height=650,
    width=1200,
    legend_title="Metrics",
)
```

```python
# Improving Axis & Hover Style

# X-axis
fig.update_xaxes(
    title_text="Category",
    tickangle=-30
)

# Left Y-axis
fig.update_yaxes(
    title_text="Average Rating",
    secondary_y=False
)

# Right Y-axis
fig.update_yaxes(
    title_text="Total Reviews",
    separatethousands=True,
    secondary_y=True
)
```

```python
fig.show(renderer="browser")
```

### Task-2: Choropleth Map

```python
t2_df = pd.read_csv("../Data/Task_wise_dataset/task2_grouped.csv")
t2_df
```

```python
# Dummy Countries
t2_df["Country"] = [
    "United States",
    "Russia",          
    "Brazil",  
    "India",          
    "Canada"  
]
```

```python
# Highlight Columns
t2_df["Highlight"] = np.where(
    t2_df["Installs"] > 1_000_000,
    "High Install",
    "Normal"
)
```

```python
t2_df
```

```python
# Choropleth Map
fig = px.choropleth(
    t2_df,
    locations="Country",
    locationmode="country names",
    color="Installs",
    hover_name="Category",
    hover_data={
        "Installs": ":,",
        "Highlight": True
    },
    color_continuous_scale=[
        GOOGLE_COLORS["blue"],
        GOOGLE_COLORS["green"],
        GOOGLE_COLORS["yellow"],
        GOOGLE_COLORS["red"]
    ],
    title=
    "Global Installs by App Category"
)
```

```python
# Hover & Layout
fig.update_traces(
    hovertemplate=
    "<b>Country:</b> %{location}<br>" +
    "<b>Category:</b> %{hovertext}<br>" +
    "<b>Total Installs:</b> %{z:,}<br>" +
    "<b>Status:</b> %{customdata[1]}<extra></extra>"
)
```

```python
# Layout Styling
fig.update_layout(
    template="plotly_white",
    height=650,
    width=1200,
    title={
        "x": 0.5
    },
    geo=dict(
        showframe=False,
        showcoastlines=True,
        projection_type="equirectangular"
    ),
    coloraxis_colorbar=dict(
        title="Installs"
    )
)
```

```python
fig.show(renderer="browser")
```
